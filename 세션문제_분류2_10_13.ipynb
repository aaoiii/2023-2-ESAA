# 1. Lightgbm

* 데이터 '와인품질분류.zip'입니다.

1. 필요한 라이브러리를 가져옵니다.

```python
import pandas as pd
import lightgbm as lgbm
```

2. 데이터를 불러옵니다.

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
smpl_sub = pd.read_csv('sample_submission.csv')
```

```python
train.head()
```

```python
test.head()
```

3. 변수를 변환합니다.

```python
# type에는 white와 red 두 종류가 있습니다.
# 각각 0,1로 변환합니다.

train=train.replace({'type':'white'},0)
train=train.replace({'type':'red'},1)

test=test.replace({'type':'white'},0)
test=test.replace({'type':'red'},1)
```

```python
train.head()
```

```python
test.head()
```

```python
# 모델에 입력하기 전 데이터를 정형합니다.
# train_x : train의 'index', 'quality' 피처를 제거하고 저장합니다.
# train_y : train의 'quality'를 저장합니다.
# test_x : test의 'index'를 제거하고 저장합니다.


train_x=train.drop(['index','quality'],axis=1)
train_y=train['quality']
test_x=test.drop(['index'],axis=1)


```

```python
train_x.shape, train_y.shape, test_x.shape
```

```python
train_x.head()
```

```python
train_y.head()
```

```python
test_x.head()
```

4. LightGBM 모델을 생성 및 훈련합니다.

```python
lgbmmodel=lgbm.LGBMClassifier()
lgbmmodel.fit(train_x,train_y)
```

5. 생성한 모델로 예측 데이터를 생성합니다.

```python
y_pred = lgbmmodel.predict(test_x)
```

6. 제출파일을 생성합니다.

```python
smpl_sub.head()
```

```python
#서브미션 파일의 quality 변수에 예측값을 대입합니다.
```

```python
smpl_sub['quality'] = y_pred
```

```python
smpl_sub.head()
```

```python
# csv 파일로 저장합니다.
smpl_sub.to_csv('submission.csv', index=False)
```

# 2. XGBoost

```python
# 데이터 다운로드 링크로 데이터를 코랩에 불러옵니다.

!wget 'https://bit.ly/3i4n1QB'

import zipfile
with zipfile.ZipFile('3i4n1QB', 'r') as existing_zip:
    existing_zip.extractall('data')
```

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
from xgboost import XGBClassifier

train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')

train.drop('index',axis = 1 ,inplace =True)
test.drop('index',axis = 1 ,inplace =True)
```

```python
# 데이터 확인
train.head()
```

```python
test.head()
```

```python
# 원핫 인코딩 (pd.get_dummies())
ohe_train=pd.get_dummies(train)
ohe_test=pd.get_dummies(test)
```

```python
ohe_train.head()
```

```python
# 모델 정의
model=XGBClassifier()
```

```python

# LabelEncoding
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
ohe_train['quality'] = encoder.fit_transform(ohe_train['quality'])

ohe_train.head()

#quality 5>2로 변화!
```

```python
# 모델 학습
# X 는 train에서 quality 를 제외한 모든 변수
# y 는 train의 quality 변수

X=ohe_train.drop('quality',axis=1)
y=ohe_train['quality']


```

```python
# fit 메소드를 이용해 모델 학습
model.fit(X,y)
```

```python
# predict 메소드와 원핫 인코딩한 test 데이터를 이용해 품질 예측
pred = model.predict(ohe_test)
```

```python
# sample_submission.csv 파일을 불러와 예측된 값으로 채워 주기
submission = pd.read_csv('data/sample_submission.csv')
submission['quality'] = pred + 3
submission.head()
```

```python
# 정답파일 내보내기
submission.to_csv('xgb_pred.csv',index = False)
```

# 3. 보팅

```python
# 데이터 다운로드 링크로 데이터를 코랩에 불러옵니다.

!wget 'https://bit.ly/3i4n1QB'

import zipfile
with zipfile.ZipFile('3i4n1QB', 'r') as existing_zip:
    existing_zip.extractall('data')
```

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier

from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier

train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')

train.drop('index',axis = 1 ,inplace =True)
test.drop('index',axis = 1 ,inplace =True)
```

```python
# 데이터 확인
train.head()
```

```python
# 원핫 인코딩 (pd.get_dummies())

train_ohe = pd.get_dummies(train)
test_ohe = pd.get_dummies(test)
```

```python
train_ohe.head()
```

```python
# 모델 정의 (LightGBM, XGBoost, Random Forest)
LGBM = LGBMClassifier()
XGB = XGBClassifier()
RF = RandomForestClassifier()

# VotingClassifier 정의
VC = VotingClassifier(estimators=[('rf',RF),('xgb',XGB),('lgbm',LGBM)],voting = 'soft')
```

```python
# 모델 학습

# X 는 train에서 quality 를 제외한 모든 변수
# y 는 train의 qulity 변수
X = train_ohe.drop('quality',axis= 1)
y = train_ohe['quality']


# fit 메소드를 이용해 모델 학습
VC.fit(X,y)
```

```python
# predict 메소드와 원핫 인코딩한 테스트 데이터를 이용해 품질 예측
VC.predict(test_ohe)
```
